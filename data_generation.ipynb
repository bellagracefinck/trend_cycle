```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
import requests
from bs4 import BeautifulSoup
import time
```

```python
# creation of 2000s-2020s SSYY links

years = [i for i in range(2000,2024)]
seasons = ['spring','fall']
season_links = []

for year in years:
    for season in seasons:
        season_links.append('https://www.vogue.com/fashion-shows/'+season+'-'+str(year)+'-ready-to-wear')
```

```python
# creation of older (1990-1999) SSYY links

years = [i for i in range(1990,2000)]
seasons = ['spring','fall']
older_season_links = []

for year in years:
    for season in seasons:
        older_season_links.append('https://www.vogue.com/fashion-shows/'+season+'-'+str(year)+'-ready-to-wear')
```

```python
import requests
import time
from bs4 import BeautifulSoup
start_time = time.time() 
link_dict = {}
for season in season_links:
    print(season)
    # Season URL
    URL = season
    
    # Empty lists for names and links
    temp_links = []

    # Fetch the webpage
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
   
    try: 
        # Find all elements with the specified class
        elements = soup.find("div", class_='GroupedNavigationLinks-igsvOK iBwhBc')

        # Find all <a> tags within the element
        links = elements.find_all('a', class_='NavigationInternalLink-cWEaeo knabMb grouped-navigation__link link--secondary navigation__link')

        # Extract designer names and links
        for link in links:
            designer_name = link.text.strip()  
            designer_link = link['href'] 
            #print("Designer:", designer_name)
            #print("Link:", designer_link)
            temp_links.append(designer_link)

        link_dict[str(season[36:47])] = temp_links
    except AttributeError:
        print("No navigation links found.")
    
end_time = time.time()  # Record the end time
execution_time = end_time - start_time  # Calculate execution time
print("Execution time:", execution_time, "seconds")
```

```python
start_time = time.time() 
link_dict2 = {}
for season in older_season_links:
    print(season)
    # Season URL
    URL = season
    
    # Empty lists for names and links
    temp_links = []

    # Fetch the webpage
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
   
    try: 
        # Find all elements with the specified class
        elements = soup.find("div", class_='GroupedNavigationLinks-igsvOK iBwhBc')

        # Find all <a> tags within the element
        links = elements.find_all('a', class_='NavigationInternalLink-cWEaeo knabMb grouped-navigation__link link--secondary navigation__link')

        # Extract designer names and links
        for link in links:
            designer_name = link.text.strip()  
            designer_link = link['href'] 
            #print("Designer:", designer_name)
            #print("Link:", designer_link)
            temp_links.append(designer_link)

        link_dict2[str(season[36:47])] = temp_links
    except AttributeError:
        print("No navigation links found.")
    
end_time = time.time()  # Record the end time
execution_time = end_time - start_time  # Calculate execution time
print("Execution time:", execution_time, "seconds")
```

```python
# function to add season from link

def add_season(key, link_dict):
    output = pd.DataFrame(columns = ['season','year','seasonyear', 'designer','text'])
    start_time = time.time() 
    suffixes = link_dict[key]
    
    x = suffixes[0].split('/')[2].split('-')
    season = x[0]
    year = x[1]
    SYY = str(season)[0]+str(year)
    print(SYY)
    
    for i in range(len(suffixes)):
        URL = 'https://www.vogue.com'+suffixes[i]
        designer_name = suffixes[i].split('/')[3]
        #print(designer_name)
        page = requests.get(URL)
        soup = BeautifulSoup(page.content, "html.parser")
        review = soup.find('div', class_ = 'body__inner-container')
        
        rev = str(review)[38:]
        
        output.loc[len(output)+1] = {'season':season, 'year':year, 'seasonyear':SYY, 'designer': designer_name, 'text' :rev}
    
    
    end_time = time.time()  # Record the end time
    execution_time = end_time - start_time  # Calculate execution time
    print("Execution time:", execution_time, "seconds")
    
    return output
```

### Individuals mentioned in reviews
Frequently, designers, models, and creative directors are mentioned by name in the collection reviews. As this can unfairly inform supervised learning algorithms, we must identify them and keep track of the people mentioned in all of the reviews so we are able to filter them out if necessary. 

I utilized the StanfordNERTagger to identify people and wrote a .txt file (available at https://github.com/bellagracefinck/trend_cycle/blob/main/people.txt) to keep track.

```python
data = pd.read_csv('complete.csv').drop(columns = ['Unnamed: 0.4', 'Unnamed: 0.3','Unnamed: 0.2','Unnamed: 0.1'])
data['id'] = data.index

data = data.fillna('drop')
data = data[data['text']!= 'drop']

data['raw_tokens'] = [word_tokenize(sentence) for sentence in data['text']]
sample_tokens = (data[['id','raw_tokens']].explode('raw_tokens'))
sample_tokens.head()
```

```python
# creating list of people (proper nouns) mentioned in all collection reviews
from nltk.tag import StanfordNERTagger
import time
from nltk import pos_tag, word_tokenize

jar = 'stanford/stanford-ner-4.2.0.jar'
model = 'stanford/classifiers/english.all.3class.distsim.crf.ser.gz'

st = StanfordNERTagger(model, jar) 
people = []

start = time.time()

text = list(sample_tokens['raw_tokens'].unique())
people += [word[0] for word in st.tag(text) if word[1] == 'PERSON']

end = time.time()
print((end-start)/60)
```

```python
# create list of people's names that show up in the dataset
with open("people.txt", "w") as output:
    output.write(str(people))
```
